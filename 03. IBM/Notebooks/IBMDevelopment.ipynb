```python
data = pd.read_csv('../Data/IBMCustomerChurn.csv')
data.head()
```

```python
data.info()
```

```python
other_services = data.iloc[:,[9,10,11,12,13,14]].copy()
other_services['OtherServices'] = 0
other_services
```

```python
for i in list(range(0,other_services.shape[0])):
    boolean = (other_services.loc[i,'OnlineSecurity'] != 'No') | (other_services.loc[i,'OnlineBackup'] != 'No'
        ) | (other_services.loc[i,'DeviceProtection'] != 'No') | (other_services.loc[i,'TechSupport'] != 'No'
        ) | (other_services.loc[i,'StreamingTV'] != 'No') | (other_services.loc[i,'StreamingMovies'] != 'No')
    if boolean == True:
        other_services.loc[i,'OtherServices'] = 1
        
other_services
```

```python
other_services = pd.get_dummies(other_services)
other_services.head()
```

```python
other_services = other_services.drop(columns=['OtherServices', 'OnlineSecurity_No', 'OnlineSecurity_No internet service', 'OnlineBackup_No', 'OnlineBackup_No internet service', 'DeviceProtection_No', 'DeviceProtection_No internet service', 'TechSupport_No', 'TechSupport_No internet service', 'StreamingTV_No', 'StreamingTV_No internet service', 'StreamingMovies_No', 'StreamingMovies_No internet service'])
other_services
```
